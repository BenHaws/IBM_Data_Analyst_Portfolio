Scrape the `Language name` and `annual average salary`.


```python
from bs4 import BeautifulSoup
import requests

# URL of the webpage to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"

# Download the webpage content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Find all table rows (tr) in the table body (tbody)
    table_rows = soup.find("tbody").find_all("tr")

    # Initialize lists to store language names and annual average salaries
    language_names = []
    average_annual_salary = []

    # Iterate over each table row
    for row in table_rows:
        # Find all table data (td) elements in the row
        cells = row.find_all("td")
        # Extract the language name from the first cell (index 0)
        language_names.append(cells[0].text.strip())
        # Extract the annual average salary from the second cell (index 1)
        average_annual_salary.append(cells[1].text.strip())

    # Print the scraped data
    for language, salary in zip(language_names, average_annual_salary):
        print(f"Language: {language}, Average Annual Salary: {salary}")

else:
    print("Failed to retrieve webpage.")
```
